```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("campeonato-brasileiro-gols.csv")
```

```python
list = ['Santos', 'Sao Paulo', 'Corinthians', 'Palmeiras']
```

```python
data = data.loc[data['clube'].isin(list)]
```

```python
data.head()
```

```python
#convertendo tempo
for i, value in data.iterrows():
    valor = value['minuto']
    if "+" in str(valor):
        num = valor.split("+")
        soma = 0
        for n in num:
            soma += int(n)
        data.at[i,'minuto'] = soma
```

```python
data.dtypes
```

```python
#convertendo tipo
data['minuto'] = data['minuto'].astype(str).astype(int)
data['rodata'] = data['rodata'].astype(str).astype(int)
```

```python
data.dtypes
```

```python
sns.displot(data = data, x = 'minuto', hue = 'clube', kind = 'hist', kde= True)
plt.title('Distribuição Gols')
plt.ylabel('Gols')
plt.xlabel('Minutagem')
```
